```python
def exibir_menu():
    print("\n--- MINHA AGENDA DI츼RIA ---")
    print("1. Adicionar Tarefa")
    print("2. Visualizar Agenda")
    print("3. Marcar Tarefa como Conclu칤da")
    print("4. Sair")

def mostrar_agenda(agenda):
    if not agenda:
        print("\nSua agenda est치 vazia!")
    else:
        print("\n--- COMPROMISSOS ---")
        for i, tarefa in enumerate(agenda):
            status = "九" if tarefa['concluida'] else "仇"
            print(f"{i + 1}. [{tarefa['hora']}] {tarefa['descricao']} - {status}")

def principal():
    agenda = []

    while True:
        exibir_menu()
        opcao = input("Escolha uma op칞칚o: ")

        if opcao == "1":
            hora = input("Hor치rio (ex: 09:00): ")
            descricao = input("O que precisa fazer? ")
            agenda.append({"hora": hora, "descricao": descricao, "concluida": False})
            print("Tarefa adicionada!")

        elif opcao == "2":
            mostrar_agenda(agenda)

        elif opcao == "3":
            mostrar_agenda(agenda)
            if agenda:
                indice = int(input("Digite o n칰mero da tarefa conclu칤da: ")) - 1
                if 0 <= indice < len(agenda):
                    agenda[indice]['concluida'] = True
                    print("Status atualizado!")
                else:
                    print("N칰mero inv치lido.")

        elif opcao == "4":
            print("Saindo... Tenha um 칩timo dia!")
            break
        else:
            print("Op칞칚o inv치lida, tente novamente.")

if __name__ == "__main__":
    principal()
```

```python
import streamlit as st

# Configura칞칚o da p치gina
st.set_page_config(page_title="Minha Agenda Di치ria", page_icon="游늰")
st.title("游늰 Minha Agenda Di치ria")

# Inicializa a agenda na mem칩ria do navegador
if 'agenda' not in st.session_state:
    st.session_state.agenda = []

# --- Barra Lateral para Adicionar Tarefas ---
st.sidebar.header("Nova Tarefa")
hora = st.sidebar.text_input("Hor치rio (ex: 09:00)")
descricao = st.sidebar.text_input("O que precisa fazer?")

if st.sidebar.button("Adicionar"):
    if hora and descricao:
        st.session_state.agenda.append({"hora": hora, "descricao": descricao, "concluida": False})
        st.success("Tarefa adicionada!")
    else:
        st.error("Preencha todos os campos.")

# --- Exibi칞칚o da Agenda ---
st.subheader("Compromissos do Dia")

if not st.session_state.agenda:
    st.info("Sua agenda est치 vazia!")
else:
    for i, tarefa in enumerate(st.session_state.agenda):
        col1, col2, col3 = st.columns([1, 4, 1])

        status = "九" if tarefa['concluida'] else "仇"
        col1.write(f"**{tarefa['hora']}**")
        col2.write(f"{tarefa['descricao']} ({status})")

        if col3.button("Ok", key=f"btn_{i}"):
            st.session_state.agenda[i]['concluida'] = True
            st.rerun()

if st.button("Limpar Agenda"):
    st.session_state.agenda = []
    st.rerun()
```
